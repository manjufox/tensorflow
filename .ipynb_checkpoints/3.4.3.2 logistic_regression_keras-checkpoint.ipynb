```python
import numpy as np
```

```python
from keras.models import Sequential
```

```python
from keras.layers import Dense, Activation
```

```python
from keras.optimizers import SGD
```

```python
#Sequential()は層構造のモデルを定義するためのメソッド
model = Sequential([
    Dense(input_dim=2, units=1), # w1x1 + w2x2 + b と同じ
    Activation('sigmoid') #　活性化関数
])
```

```python
#箇条書きで書くことも可能
model = Sequential()
model.add(Dense(input_dim=2, units=1))
model.add(Activation('sigmoid'))
```

```python
#確率的勾配降下法
model.compile(loss='binary_crossentropy', optimizer=SGD(lr=0.1)) # lrは学習率(learning rate)
```

```python
#データ用意
X = np.array([[0,0],[0,1],[1,0],[1,1]])
Y = np.array([[0],[1],[1],[1]])
```

```python
#学習実行
#他のプロセス(jupyter)が実行中だとエラーが出る
model.fit(X,Y,epochs=200,batch_size=1)
```

```python
#結果
classes = model.predict_classes(X,batch_size=1)
prob = model.predict_proba(X,batch_size=1)
```

```python
#結果表示
print('classified:')
print(Y == classes)
print()
print('output probability:')
print(prob)
```
